```python
import matplotlib.pyplot as plt
import numpy as np
import random


def random_walk_2d(num_steps):
    x, y = 0, 0
    x_positions = [x]
    y_positions = [y]

    for _ in range(num_steps):
        angle = random.uniform(0, 2 * np.pi)
        x += np.cos(angle)
        y += np.sin(angle)
        x_positions.append(x)
        y_positions.append(y)

    return x_positions, y_positions

def plot_random_walk(x_positions, y_positions):
    plt.figure(figsize=(8, 8))
    plt.plot(x_positions, y_positions, marker='o', markersize=2, linewidth=1)
    plt.title('2D Random Walk')
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    plt.axis('equal')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    num_steps = 1000
    x_positions, y_positions = random_walk_2d(num_steps)
    plot_random_walk(x_positions, y_positions)
```

```python
def plot_random_numbers_distribution():
    random_numbers = [random.randint(1, 6) for _ in range(10000000)]
    distribution = {i: random_numbers.count(i) for i in range(1, 7)}
    plt.bar(distribution.keys(), distribution.values())
    plt.title("Random Numbers Distribution (1-6)")
    plt.xlabel("Number")
    plt.ylabel("Frequency")
    plt.xticks(range(1, 7))
    plt.show()

plot_random_numbers_distribution()
```
