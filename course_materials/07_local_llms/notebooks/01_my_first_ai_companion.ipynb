```python
# Let's install our ninja tools! ğŸ¥·
import subprocess
import sys

def install_package(package):
    """Install a Python package - like learning a new jutsu!"""
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        return True
    except:
        return False

# Our essential ninja tools
ninja_tools = {
    'requests': 'For talking to Ollama ğŸ“¡',
    'ipywidgets': 'For creating cool interfaces ğŸ¨',
    'matplotlib': 'For drawing charts ğŸ“Š',
    'colorama': 'For colorful text ğŸŒˆ'
}

print("ğŸ¥· Installing ninja tools...\n")

for tool, description in ninja_tools.items():
    try:
        __import__(tool)
        print(f"âœ… {tool} - {description} (Already installed!)")
    except ImportError:
        print(f"ğŸ“¦ Installing {tool} - {description}")
        if install_package(tool):
            print(f"âœ… {tool} installed successfully!")
        else:
            print(f"âŒ Failed to install {tool}")

print("\nğŸ‰ All ninja tools ready! Let's continue!")
```

## Step 2: Creating Our AI Companion Class ğŸ¤–

Let's create a special class to talk to our AI. Think of this as creating a **summoning circle** for our AI spirits!

```python
class AICompanion:
    """Your personal AI companion - like having a digital friend! ğŸ¤–âœ¨"""
    
    def __init__(self, name="AI-chan", ollama_url="http://localhost:11434"):
        self.name = name
        self.ollama_url = ollama_url
        self.conversation_history = []
        self.personality = "friendly and helpful"
        
        print(f"ğŸŒŸ {self.name} is awakening...")
        
    def check_ollama_status(self):
        """Check if Ollama is running - like checking if the summoning circle is active!"""
        try:
            response = requests.get(f"{self.ollama_url}/api/tags", timeout=5)
            if response.status_code == 200:
                print("âœ… Ollama is running! The summoning circle is active! ğŸ”®")
                return True
            else:
                print("âŒ Ollama is not responding properly")
                return False
        except requests.exceptions.RequestException:
            print("âŒ Ollama is not running! Please start Ollama first.")
            print("ğŸ’¡ Tip: Visit https://ollama.ai to download and install Ollama")
            return False
    
    def list_available_models(self):
        """List all AI models we can summon! ğŸ“œ"""
        try:
            response = requests.get(f"{self.ollama_url}/api/tags")
            if response.status_code == 200:
                models = response.json().get('models', [])
                if models:
                    print(f"ğŸ­ Available AI spirits ({len(models)} found):")
                    for i, model in enumerate(models, 1):
                        name = model.get('name', 'Unknown')
                        size = model.get('size', 0) / (1024**3)  # Convert to GB
                        print(f"   {i}. {name} ({size:.1f} GB)")
                    return [model['name'] for model in models]
                else:
                    print("ğŸ“­ No AI models found! Let's download one!")
                    return []
            else:
                print("âŒ Could not fetch model list")
                return []
        except Exception as e:
            print(f"âŒ Error: {e}")
            return []
    
    def download_model(self, model_name):
        """Download a new AI model - like catching a new PokÃ©mon! ğŸ¯"""
        print(f"ğŸ“¥ Summoning {model_name}... This might take a while!")
        print("â˜• Perfect time for a coffee break!")
        
        try:
            response = requests.post(
                f"{self.ollama_url}/api/pull",
                json={"name": model_name},
                stream=True
            )
            
            for line in response.iter_lines():
                if line:
                    data = json.loads(line.decode('utf-8'))
                    status = data.get('status', '')
                    
                    if 'pulling' in status.lower():
                        print(f"ğŸ”„ {status}")
                    elif 'success' in status.lower():
                        print(f"ğŸ‰ {model_name} successfully summoned!")
                        return True
                        
            return True
        except Exception as e:
            print(f"âŒ Failed to summon {model_name}: {e}")
            return False
    
    def chat(self, message, model="llama2:7b-chat", system_prompt=None):
        """Chat with your AI companion! ğŸ’¬"""
        if system_prompt is None:
            system_prompt = f"You are {self.name}, a {self.personality} AI assistant."
        
        try:
            print(f"ğŸ¤” {self.name} is thinking...")
            
            payload = {
                "model": model,
                "prompt": message,
                "system": system_prompt,
                "stream": False
            }
            
            start_time = time.time()
            response = requests.post(
                f"{self.ollama_url}/api/generate",
                json=payload
            )
            end_time = time.time()
            
            if response.status_code == 200:
                result = response.json()
                ai_response = result.get('response', 'Sorry, I had trouble responding.')
                
                # Store conversation
                self.conversation_history.append({
                    'timestamp': datetime.now(),
                    'user': message,
                    'ai': ai_response,
                    'model': model,
                    'response_time': end_time - start_time
                })
                
                print(f"ğŸ’¬ {self.name}: {ai_response}")
                print(f"â±ï¸ Response time: {end_time - start_time:.2f} seconds")
                
                return ai_response
            else:
                print(f"âŒ Error: {response.status_code} - {response.text}")
                return None
                
        except Exception as e:
            print(f"âŒ Chat error: {e}")
            return None
    
    def set_personality(self, personality):
        """Change your AI's personality! ğŸ­"""
        self.personality = personality
        print(f"ğŸ­ {self.name} personality changed to: {personality}")
    
    def get_conversation_stats(self):
        """Get stats about your conversations! ğŸ“Š"""
        if not self.conversation_history:
            print("ğŸ“­ No conversations yet!")
            return
        
        total_conversations = len(self.conversation_history)
        avg_response_time = sum(c['response_time'] for c in self.conversation_history) / total_conversations
        
        print(f"ğŸ“Š Conversation Stats for {self.name}:")
        print(f"   ğŸ’¬ Total conversations: {total_conversations}")
        print(f"   â±ï¸ Average response time: {avg_response_time:.2f} seconds")
        print(f"   ğŸ• First chat: {self.conversation_history[0]['timestamp'].strftime('%Y-%m-%d %H:%M:%S')}")
        print(f"   ğŸ• Last chat: {self.conversation_history[-1]['timestamp'].strftime('%Y-%m-%d %H:%M:%S')}")

# Create your first AI companion!
ai_friend = AICompanion(name="Sakura-chan")
print("ğŸŒ¸ Sakura-chan is ready to be your AI companion!")
```

## Step 4: Downloading Your First AI Model ğŸ“¥

If you don't have any models yet, let's download a small, friendly one! This is like catching your first PokÃ©mon! ğŸ¯

```python
# Let's create an interactive model downloader!
def create_model_downloader():
    """Create a fun interface to download AI models!"""
    
    # Beginner-friendly models
    beginner_models = {
        "ğŸ¦™ Llama 2 Chat (7B) - Great for conversations!": "llama2:7b-chat",
        "ğŸ”¥ Mistral (7B) - Fast and smart!": "mistral:7b",
        "âš¡ Phi-2 (2.7B) - Tiny but mighty!": "phi:2.7b",
        "ğŸ¤– Code Llama (7B) - Perfect for coding help!": "codellama:7b"
    }
    
    model_dropdown = widgets.Dropdown(
        options=list(beginner_models.keys()),
        value=list(beginner_models.keys())[0],
        description='Choose Model:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='500px')
    )
    
    download_button = widgets.Button(
        description='ğŸ“¥ Download Model',
        button_style='success',
        icon='download',
        layout=widgets.Layout(width='200px')
    )
    
    output_area = widgets.Output()
    
    def on_download_click(b):
        with output_area:
            clear_output()
            selected_model = beginner_models[model_dropdown.value]
            print(f"ğŸ¯ Starting download of {selected_model}...")
            print("â³ This might take 5-15 minutes depending on your internet speed")
            print("â˜• Perfect time for a snack break!")
            
            success = ai_friend.download_model(selected_model)
            
            if success:
                print(f"\nğŸ‰ Success! {selected_model} is now ready!")
                print("ğŸš€ You can now chat with your AI companion!")
            else:
                print(f"\nğŸ˜… Something went wrong downloading {selected_model}")
                print("ğŸ’¡ Try checking your internet connection and try again!")
    
    download_button.on_click(on_download_click)
    
    return widgets.VBox([
        widgets.HTML("<h3>ğŸ¯ Download Your First AI Model!</h3>"),
        widgets.HTML("<p>Choose a model to download. Don't worry, you can always download more later!</p>"),
        model_dropdown,
        download_button,
        output_area
    ])

# Only show downloader if we need models
if ollama_status:
    current_models = ai_friend.list_available_models()
    if not current_models:
        display(create_model_downloader())
    else:
        print("âœ… You already have models! Let's start chatting!")
else:
    print("âš ï¸ Please install and start Ollama first!")
```

## Step 6: Interactive Chat Interface! ğŸ®

Now let's create a cool interactive chat interface where you can talk to your AI companion anytime! It's like having a messaging app with AI! ğŸ“±

```python
def create_interactive_chat():
    """Create an awesome interactive chat interface!"""
    
    # Get available models
    available_models = ai_friend.list_available_models()
    
    if not available_models:
        return widgets.HTML("<p style='color: red;'>âŒ No models available! Please download a model first.</p>")
    
    # Create widgets
    model_selector = widgets.Dropdown(
        options=available_models,
        value=available_models[0],
        description='AI Model:',
        style={'description_width': 'initial'}
    )
    
    personality_selector = widgets.Dropdown(
        options=[
            "ğŸŒ¸ Friendly and cheerful (like Sakura from Naruto)",
            "âš¡ Energetic and determined (like Naruto)",
            "ğŸ§  Calm and wise (like Kakashi)",
            "ğŸ˜Š Bubbly and optimistic (like Hinata)",
            "ğŸ”¥ Passionate and strong (like Natsu from Fairy Tail)",
            "â„ï¸ Cool and collected (like Todoroki from MHA)"
        ],
        value="ğŸŒ¸ Friendly and cheerful (like Sakura from Naruto)",
        description='Personality:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='400px')
    )
    
    message_input = widgets.Textarea(
        placeholder="Type your message here... (e.g., 'Tell me about yourself!' or 'What's your favorite anime?')",
        description='Message:',
        layout=widgets.Layout(width='100%', height='80px'),
        style={'description_width': 'initial'}
    )
    
    send_button = widgets.Button(
        description='ğŸ’¬ Send Message',
        button_style='primary',
        icon='paper-plane'
    )
    
    clear_button = widgets.Button(
        description='ğŸ—‘ï¸ Clear Chat',
        button_style='warning',
        icon='trash'
    )
    
    stats_button = widgets.Button(
        description='ğŸ“Š Show Stats',
        button_style='info',
        icon='chart-bar'
    )
    
    chat_output = widgets.Output()
    
    # System prompts for different personalities
    personality_prompts = {
        "ğŸŒ¸ Friendly and cheerful (like Sakura from Naruto)": "You are Sakura-chan, a cheerful and friendly AI assistant inspired by Sakura Haruno from Naruto. You're helpful, determined, and always encouraging. You occasionally use phrases like 'Believe it!' and 'You can do it!' Use emojis to express yourself.",
        "âš¡ Energetic and determined (like Naruto)": "You are an energetic AI assistant inspired by Naruto Uzumaki. You're super enthusiastic, never give up, and always believe in people. You love ramen and use phrases like 'Dattebayo!' and 'I'll never give up!' You're optimistic and encouraging.",
        "ğŸ§  Calm and wise (like Kakashi)": "You are a calm and wise AI assistant inspired by Kakashi Hatake. You're knowledgeable, patient, and give thoughtful advice. You occasionally make references to books and speak in a measured, wise manner.",
        "ğŸ˜Š Bubbly and optimistic (like Hinata)": "You are a gentle and kind AI assistant inspired by Hinata Hyuga. You're shy but caring, always supportive, and speak softly. You believe in people and are very encouraging, though sometimes a bit timid.",
        "ğŸ”¥ Passionate and strong (like Natsu from Fairy Tail)": "You are a passionate AI assistant inspired by Natsu Dragneel from Fairy Tail. You're fired up, loyal to friends, and always ready for action. You use phrases like 'I'm all fired up!' and talk about the power of friendship.",
        "â„ï¸ Cool and collected (like Todoroki from MHA)": "You are a cool and analytical AI assistant inspired by Shoto Todoroki from My Hero Academia. You're calm, logical, and speak in a measured way. You're helpful but maintain a composed demeanor."
    }
    
    def send_message(b):
        if not message_input.value.strip():
            return
        
        with chat_output:
            user_message = message_input.value
            selected_model = model_selector.value
            selected_personality = personality_selector.value
            system_prompt = personality_prompts[selected_personality]
            
            print(f"\n{'='*60}")
            print(f"ğŸ‘¤ You: {user_message}")
            print(f"ğŸ¤– {ai_friend.name} is thinking...")
            
            response = ai_friend.chat(
                user_message,
                model=selected_model,
                system_prompt=system_prompt
            )
            
            if not response:
                print("ğŸ˜… Sorry, I had trouble responding. Please try again!")
        
        # Clear the input
        message_input.value = ""
    
    def clear_chat(b):
        ai_friend.conversation_history = []
        chat_output.clear_output()
        with chat_output:
            print("ğŸ—‘ï¸ Chat history cleared! Ready for a fresh start!")
    
    def show_stats(b):
        with chat_output:
            print(f"\n{'='*60}")
            ai_friend.get_conversation_stats()
    
    # Connect button events
    send_button.on_click(send_message)
    clear_button.on_click(clear_chat)
    stats_button.on_click(show_stats)
    
    # Create the interface
    return widgets.VBox([
        widgets.HTML("<h3>ğŸ® Interactive AI Chat Interface</h3>"),
        widgets.HTML("<p>Choose your AI's personality and start chatting! Try asking about anime, coding, or anything else!</p>"),
        widgets.HBox([model_selector, personality_selector]),
        message_input,
        widgets.HBox([send_button, clear_button, stats_button]),
        chat_output
    ])

# Display the interactive chat
if ollama_status:
    display(create_interactive_chat())
else:
    print("âš ï¸ Please start Ollama first to use the interactive chat!")
```
