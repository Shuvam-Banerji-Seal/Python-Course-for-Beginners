```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

# Multiple Lines in a Single Plot

```python
x = [0, 1, 2, 3, 4]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.plot(x, y1, label="y = x²", color='blue')
plt.plot(x, y2, label="y = x³", color='green')
plt.title("Multiple Lines")
plt.legend()
plt.grid(True)
plt.show()
```

# Scatter Plots

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
print(type(x), type(y))  # This will print the types of x and y
colors = np.random.rand(50)

plt.scatter(x, y, c=colors, cmap='viridis', s=100, alpha=0.7)
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```

# Bar Chart

# Histogram


```python
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple', alpha=0.7, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
```

# Box Plot


```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True)
plt.title("Box Plot")
plt.xlabel("Distribution")
plt.xticks([1, 2, 3], ['Std 1', 'Std 2', 'Std 3'])
plt.ylabel("Value")
```

# Subplots 


```python
fig, axs = plt.subplots(2, 2)

x = [0, 1, 2, 3]

axs[0, 0].plot(x, [i**2 for i in x])
axs[0, 1].bar(x, [i + 1 for i in x])
axs[1, 0].scatter(x, [i - 1 for i in x])
axs[1, 1].hist([i for i in x]*5)

plt.tight_layout()
plt.show()
```

# Annotations


```python
# Annotations

x = np.linspace(0, 10, 100)
print(x)  # This will print the type of x
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave with Annotation")
plt.annotate("Peak", xy=(1.57, 1), xytext=(2.3, -0.2),
             arrowprops=dict(facecolor='blue', shrink=1))
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = 2 * x + 1

plt.plot(x, y, label='f(x) = 2x + 1')
plt.title("Linear Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
y = x**2 - 4*x + 3

plt.plot(x, y, label='f(x) = x² - 4x + 3', color='orange')
plt.title("Quadratic Function")
plt.grid(True)
plt.legend()
plt.show()
```

```python
y = x**3

plt.plot(x, y, label='f(x) = x³', color='purple')
plt.title("Cubic Function")
plt.grid(True)
plt.legend()
plt.show()
```

```python
y = np.exp(x)

plt.plot(x, y, label='f(x) = e^x', color='green')
plt.ylim(0, 5000)  # To avoid scale issues
plt.title("Exponential Function")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(0.1, 10, 100)
y = np.log(x)

plt.plot(x, y, label='f(x) = ln(x)', color='red')
plt.title("Logarithmic Function")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, np.tan(x), label='tan(x)', linestyle='dashed', alpha=0.6)
plt.ylim(-10, 10)
plt.title("Trigonometric Functions")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 5, 500)
y = np.exp(-x) * np.sin(2 * np.pi * x)

plt.plot(x, y, label='f(x) = e^{-x} * sin(2πx)', color='brown')
plt.title("Damped Oscillation")
plt.grid(True)
plt.legend()
plt.show()
```

## Sine wave animation

```python
fig, ax = plt.subplots()
x = np.linspace(0, 2 * np.pi, 100)
line, = ax.plot(x, np.sin(x), color='blue')

ax.set_ylim(-1.5, 1.5)
ax.set_title("Sine Wave Animation")

def update(frame):
    line.set_ydata(np.sin(x + frame / 10.0))
    print(type(line))  # This will print the type of line
    return line,

ani = animation.FuncAnimation(fig, update, frames=100, interval=50, blit=True)
plt.show()
```

# Complex Numbers

```python
z = 3 + 4j
print(type(z))  # This will print the type of z
print(f"Real part: {z.real}, Imaginary part: {z.imag}")
print(f"Complex number: {z}, Magnitude: {abs(z)}, Phase: {np.angle(z)}")
```

```python
import matplotlib.pyplot as plt

# Sample complex numbers
z = [1 + 2j, 2 - 1j, -1 + 1j, -2 - 2j]

# Separate real and imaginary parts
x = [num.real for num in z]
y = [num.imag for num in z]

plt.scatter(x, y, color='red')
for i, num in enumerate(z):
    plt.text(x[i], y[i], f'z{i+1}', fontsize=12)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.title("Complex Numbers on Argand Plane")
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of complex numbers
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y  # Complex grid

# Function f(z) = z^2
F = Z**2

# Plot |f(z)| = modulus
plt.figure(figsize=(6,5))
plt.contourf(X, Y, np.abs(F), levels=50, cmap='viridis')
plt.colorbar(label='|f(z)|')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Modulus of f(z) = z²')
plt.axis('equal')
plt.show()
```

```python
# Plot arg(f(z)) = phase of complex function
plt.figure(figsize=(6,5))
plt.contourf(X, Y, np.angle(F), levels=50, cmap='viridis')
plt.colorbar(label='arg(f(z))')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Phase of f(z) = z²')
plt.axis('equal')
plt.show()
```
